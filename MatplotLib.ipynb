Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It
provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,
WxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:
    Generally easy to get started for simple plots
    Support for custom labels and texts
    Great control of every element inafigure
    High-quality output in many formats
    Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
plt.scatter(x, y, c = 'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
x = np.arange(0, 10)
y = x*x
plt.plot(x, y)
```

```python
plt.plot(x, y, 'ro')
```

```python
plt.plot(x, y, 'r*', linestyle = 'dashed', linewidth = 2, markersize = 12)
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x1, y1)
ax2.plot(x2, y2)
```

```python
from scipy.stats import gamma
alpha, N = 2, 10000

# generate random numbers and plot histogram
r = gamma.rvs(alpha, size=N)
counts, bins = np.histogram(r,int(np.sqrt(N)))
plt.hist(bins[:-1], bins, weights=counts,density=True)

# plot pdf
x = np.linspace(0,15,10000)
y = gamma.pdf(x,alpha)
plt.plot(x,y)

# compute (sample) mean and median
mean = gamma.mean(alpha)
median = gamma.median(alpha)
sample_mean = np.mean(r)
sample_median = np.median(r)
print("Mean: "+str(mean))
print("Sample Mean: "+str(sample_mean))
print("Median: "+str(median))
print("Sample Median: "+str(sample_median))
plt.axvline(x=mean, color='blue', label='mean')
plt.axvline(x=median, color='red', label='median')
plt.legend()
plt.show()
```
